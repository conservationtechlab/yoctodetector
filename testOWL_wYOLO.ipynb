```python
#TEST 

#load in the model
mainDir = "/home/kyra/Desktop/OWL"
currRun = "train3"

model = YOLO(mainDir + "/runs/detect/" + currRun + "/weights/best.pt")
# model
```

```python
#save the entire set of images in the directory to a list
import os
dir_path = '/mnt/machinelearning/BUOW/100RECNX'
ImgList = []
if os.path.exists(dir_path):
    #open the file for iteration
    for filename in os.listdir(dir_path):
        if filename.lower().endswith('.jpg'):
#             print(filename)
            ImgList.append(filename)
else:
    print("Paths does not exist")
```

```python
#run the prediction on all the images in the directory
for img in ImgList:  
    FILE = dir_path + "/" + img
    results = model.predict(source=FILE, conf=0.25, save=True)
```
